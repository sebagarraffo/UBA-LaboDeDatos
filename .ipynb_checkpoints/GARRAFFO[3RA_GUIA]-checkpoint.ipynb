```python
from gapminder import gapminder
import seaborn.objects as so
```

```python
#EJ 2 
paisesAmerica = gapminder[gapminder["continent"]== "Americas"]

(
so.Plot(data = paisesAmerica, x = "year", y = "lifeExp", group = "country", color = "country")
.add(so.Line())
)
```

```python
#EJ 3

import matplotlib.pyplot as plt

paisesAmerica = gapminder[gapminder["continent"]== "Americas"]
paisesAsia = gapminder[gapminder["continent"]== "Asia"]


fig, ax =plt.subplots(1,4)  # Utilizamos este cdigo para crear una grilla de dos grficos.
fig.set_figwidth(12)

(    
so.Plot(data = paisesAmerica, x = "year", y = "lifeExp")
.add(so.Lines(color = "#bbca"), group = "country")
.add(so.Line(), so.PolyFit(1))
.on(ax[0]).plot()
)

(    
so.Plot(data = paisesAsia, x = "year", y = "lifeExp")
.add(so.Lines(color = "#bbca"), group = "country")
.add(so.Line(), so.PolyFit(1))
.on(ax[1]).plot()
)
```
