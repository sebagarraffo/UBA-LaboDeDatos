```python
#EJ 2
import seaborn as sns
penguins = sns.load_dataset("penguins")


penguins
```

```python
(
 so.Plot (
 penguins , x = "bill_length_mm", y = "bill_depth_mm" ,
 edgewidth = "body_mass_g", marker = "species" ,
 linestyle = "island", color = "species"
)
    .add(so.Dot(color = ".8"), edgecolor = "sex")
)
```

```python
(
 so.Plot(penguins, x = "bill_depth_mm", y = "bill_length_mm"
        , color = "sex")
 .add(so.Dot())
)
```

```python
(
 so.Plot(penguins, x = "bill_depth_mm", y = "bill_length_mm"
        , color = "sex")
 .add(so.Dot())
.facet("species", wrap = 3)
)
```

```python
mediaLargoPico = penguins["flipper_length_mm"].median()
promedioLargoPico = penguins["flipper_length_mm"].mean()
maxLargoPico = penguins["flipper_length_mm"].max()
minLargoPico = penguins["flipper_length_mm"].min()

print("LargoPico data:","\n", "media - ",mediaLargoPico, "\n", "promedio - ",promedioLargoPico, "\n ", "maximo - ",maxLargoPico, "\n ", "minimo - ",minLargoPico)
```
